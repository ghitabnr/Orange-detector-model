```python
# Lire le contenu du premier fichier data.yaml
with open('/content/data_combined/detection1-7/data.yaml', 'r') as file1:
    data1_content = file1.read()

# Lire le contenu du deuxième fichier data.yaml
with open('/content/data_combined/detection2-4/data.yaml', 'r') as file2:
    data2_content = file2.read()

# Lire le contenu du deuxième fichier data.yaml
with open('/content/data_combined/detection-1/data.yaml', 'r') as file3:
    data3_content = file3.read()

# Fusionner les contenus des deux fichiers en un seul
combined_content = data1_content + '\n' + data2_content + '\n' + data3_content + '\n'

# Écrire le contenu combiné dans un nouveau fichier data_combined.yaml
with open('/content/data_combined/data_combined.yaml', 'w') as file_combined:
    file_combined.write(combined_content)
```

```python
import yaml

def load_yaml(file_path):
    with open(file_path, 'r') as file:
        data = yaml.safe_load(file)
    return data

def get_classes_info(yaml_file):
    data = load_yaml(yaml_file)
    num_classes = data['nc']
    class_names = data['names']
    return num_classes, class_names

# Chemin vers votre fichier .yaml
yaml_file_path = '/content/data_combined/data_combined.yaml'

# Obtenez les informations sur les classes
num_classes, class_names = get_classes_info(yaml_file_path)

# Affichage du nombre de classes et de leurs noms
print("Nombre de classes :", num_classes)
print("Noms des classes :", class_names)
```
